```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
# tickers = yf.Tickers("TSLA AAPL MSFT GOOGL BAC MVIS NIO YMAB")

# tesla = tickers.tickers['TSLA'].history(period="max")
# tesla.to_csv('Tesla_Stocks.csv')
tesla = pd.read_csv("/home/khoi/Desktop/Stock analysis/Tesla_Stocks.csv", usecols=['Date','Open','High','Low','Close'])

tesla['Date'] = pd.to_datetime(tesla.Date,format='%Y-%m-%d')
tesla.index = tesla['Date']
tesla
```

```python
closing_prices = tesla.loc[:,'Close']
closing_prices
```

```python
plt.figure(figsize = (18,9))
plt.plot(range(tesla.shape[0]), tesla['Close'])
plt.xticks(range(0,tesla.shape[0],500),tesla['Date'].loc[::500],rotation=45)
plt.xlabel('Date',fontsize=18)
plt.ylabel('Closing Price',fontsize=18)
plt.show()
```

```python
dataset = tesla['Close'].values
new_dataset = [[i] for i in dataset]
new_dataset
```

```python
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import Dense, Dropout, LSTM

scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(new_dataset)
scaled_data    
```

```python
import numpy as np
train = tesla[0:2192]
valid = tesla[2192:]
x_train, y_train = [], []
for i in range(300,len(train)):
    x_train.append(scaled_data[i-300:i,0])
    y_train.append(scaled_data[i,0])
x_train, y_train = np.array(x_train), np.array(y_train)

x_train = np.reshape(x_train, (x_train.shape[0],x_train.shape[1],1))
```

```python
%matplotlib inline

model = Sequential()
model.add(LSTM(units=50, return_sequences=True, input_shape=(x_train.shape[1],1)))
model.add(LSTM(units=50))
model.add(Dense(1))

model.compile(loss='mean_squared_error', optimizer='adam')
model.fit(x_train, y_train, epochs=7, batch_size=1, verbose=2)

```

```python

init_inputs = tesla['Close'][len(tesla) - len(valid) - 300:].values
inputs = [[i] for i in init_inputs]

inputs = scaler.transform(inputs)

X_test = []
for i in range(300,inputs.shape[0]):
    X_test.append(inputs[i-300:i,0])
X_test = np.array(X_test)

X_test = np.reshape(X_test, (X_test.shape[0],X_test.shape[1],1))
closing_price = model.predict(X_test)
closing_price = scaler.inverse_transform(closing_price)
```

```python

valid['Predictions'] = closing_price

plt.plot(valid[['Close','Predictions']])
valid
```

To improve upon this, I will be trying to incorporate statistics to this model. I have a plan to use news as as source that should affect the results.

# Plan
Incorporate different strategies into the model or the results of the model, specifically the predictions column. One way to make this better is to calculate percent error and try to minimize this percentage. 

```python
report = valid.copy()
#positive means predictions > close
#negative means predictions < close
report['error percentage'] = (valid['Close'] - valid['Predictions'])/valid['Close'] * 100
report
```

```python
report['error percentage'].plot(figsize=(15,7))
```

```python
errorList = report['error percentage'].values.tolist()
total_negative_nums = list(filter(lambda errorList:errorList<0,errorList))
negAvg = sum(total_negative_nums)/(report['error percentage'] < 0).sum()
total_positive_nums = list(filter(lambda errorList:errorList>0,errorList))
posAvg = sum(total_positive_nums)/(report['error percentage'] > 0).sum()
```

```python
#X_test an array of length 547 with each element consisting of an array of length 300
# the first length means the days the model is trying to predict, and the second length means how many days in the past the model uses to predict
```

```python
#predicting next day
#however, the percentage error was a lot when I did validation test. I want to now make the model better after the validation test results. How do I do this?
#for now, I will use the model without the improvement

init_predict_inputs = tesla['Close'][len(tesla)-300:].values
predict_inputs = [[i] for i in init_predict_inputs]
predict_inputs = scaler.transform(predict_inputs)
X_predict = []
X_predict.append(predict_inputs[0:300,0])
    
    
X_predict = np.array(X_predict)

X_predict = np.reshape(X_predict, (X_predict.shape[0], 300, 1))
closing_price_prediction = model.predict(X_predict)
closing_price_prediction = scaler.inverse_transform(closing_price_prediction)
```

```python
#this new dataframe consists of only new predictions
#to predict the next 100 days, need to use the predicted value of today as input for next day input, and keep repeating this process.

from datetime import date
df = pd.DataFrame([ [date.today(), None, None, None, None, closing_price_prediction[0][0], None ] ], columns=['Date', 'Open', 'High', 'Low', 'Close', 'Predictions','error percentage'])
df['Low closing prediction'] = closing_price_prediction[0][0] * (1+negAvg/100)
df['High closing prediction'] = closing_price_prediction[0][0] * (1+posAvg/100)
df
```
